## Distplot

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from PIL import Image as im
import numpy as np
```

```python
a=np.random.randn(150)  # created random value to create a chart in 'distplot'
```

```python
sns.distplot(a)
```

```python
sns.distplot(a, color='green')  # canged the cart color into green
```

```python
b=pd.Series(data=a,name='Variable 1') # update data into column and named as 'Variable 1'
b
```

```python
sns.distplot(b)  # below of the chart we can see name of the column as 'Variable 1'
```

```python
sns.distplot(b, vertical=True, color='GREEN') # to change chart into vertical shape with 'green color'
```

```python
sns.distplot(b, vertical=False, color='GREEN', hist=False) # here hist is false
```

```python
sns.distplot(b, vertical=False, color='GREEN', kde=False) # here kde is False
```

```python
sns.distplot(b, vertical=False, color='GREEN', hist=False, rug=True) # here rug=True
```

## Correlation with Heatmap


`A correlation heatmap uses colored cells, typically in a monochromatic scale, to show a 2D correlation matrix (table) between two discrete dimensions or event types. It is very important in Feature Selection`

```python
df.dtypes # when we are finding co-relation data type should be float or integer.
```

```python
df.corr() # to find co-relation
```

```python
sns.heatmap(df.corr()) # it helps to visualize data in matrux format.
```

### JoinPlot


`A join plot allows to study the relationship between 2 numeric variables. The central chart display their correlation. 
It is usually a scatterplot, a hexbin plot, a 2D histogram or a 2D density plot`

### Univariate Analysis

```python
sns.jointplot(x='tip',y='total_bill',data=df,kind='hex') # 'hex' is nothing but we can see some dotted point on the chart
```

```python
sns.jointplot(x='tip',y='total_bill',data=df,kind='reg') # 'reg' will helps to draw bothh 'kde' and 'hist' and 
                                                         # regreession line also
```

```python
# above line is probability density function
```

## Pair plot

A “pairs plot” is also known as a scatterplot, in which one variable in the same data row is matched with another 
variable's value, like this: Pairs plots are just elaborations on this, showing all variables paired with all the 
other variables

```python
sns.pairplot(df) 
```

```python
sns.pairplot(df, hue='sex')
```

### Dist plot

Dist plot helps us to check the distribution of the columns feature

```python
sns.distplot(df['tip']) # if 'kde' is there it will show like 0.1, 0.2 , etc
```

```python
sns.distplot(df['tip'],kde=False,bins=10) # if 'kde' is not there it will show in count like, 1,2,3,etc.
```

## Categorical Plots

Seaborn also helps us in doing the analysis on Categorical Data points. In this section we will discuss about

1. boxplot
2. violinplot
3. countplot
4. bar plot

```python
# countplot
sns.countplot('sex', data=df)
```

```python
sns.countplot('day', data=df)
```

```python
sns.countplot(y='sex', data=df)
```

```python
## Bar plot
# it is also univariant analysis

sns.barplot(x='total_bill', y='sex', data=df) # 'bar'plot will do the same as 'countplot', but in 'bar' plot we have to
                                              # give x and y axis.
```

```python
## Bar plot

sns.barplot(y='total_bill', x='sex', data=df)
```

```python
df.head()
```

### Box plot

A box and whisker plot (sometimes called a boxplot) is a graph that presents information from a five-number summary.

```python
sns.boxplot('smoker','total_bill', data=df)
```

```python
sns.boxplot(x='day',y='total_bill', data=df, palette='rainbow')
```

```python
sns.boxplot(data=df, orient='v')
```

```python
# categorize my data based on some other categories

sns.boxplot(x='total_bill', y='day', hue='smoker', data=df)
```

## Violin Plot

`Violin plot helps us to see both the distribution of data in terms of Kernel density estimation and the box plot`

```python
sns.violinplot(x="total_bill", y="day", data=df,palette='rainbow')
```
