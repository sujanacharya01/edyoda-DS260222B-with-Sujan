### Today's class is about to create pivot table in python

```python
df1=pd.DataFrame({'A':['One','Two','Three','Four','Five'],
                 'B':['Uttam','Ajay','Pooja','Vaibhav','Usha'],
                 'C':[1,2,3,4,5],
                 'D':[10,20,30,40,50]})
```

```python
df1
```

```python
pd.pivot_table(df1, index=['A'], columns=['B'], aggfunc=np.sum, values=["D"])
```

```python
pd.pivot_table(df1, index=['A','D'], columns=['B'], aggfunc=np.sum)
```

```python
df2=pd.read_csv(r"D:\Data Science\DATA WRANGLING & VISUALIZATION\netflix_titles.csv")
df2.head(2)
```

```python
a=pd.pivot_table(df2, index=['type','date_added','listed_in'], columns=['country'], aggfunc=np.sum, values=['title'],fill_value='-')
# here 'index' will work as a 'row' function in pivot table , 'columns' will work as a 'column'
# 'aggfunc' will work as a 'values' and 'fill_value' will change the null value into whichever value is given 
# into this function
```

```python
a
```

```python
df3=pd.DataFrame({'A':['One','Two','Three','Four','Five','Six','Seven','Eight'],
                 'B':['Uttam','Ajay','Pooja','Vaibhav','Usha','Sujan','Bushan','Kiran'],
                 'C':['Large','Small','Medium','Big','Little','New','Old','Young'],
                 'D':[99,98,30,40,50,80,70,11],
                 'E':[11,37,33,21,15,99,26,45]    })
df3.head()
```

```python
pd.pivot_table(df3, index=['A','C'], aggfunc={'D':np.mean,'E':np.sum}, values=['D','E'])
# here we mentioned 'aggfunc={'D':np.mean,'E':np.sum}' to mention different values like sum, mean, etc.
```

```python
pd.pivot_table(df3, index=['A','C'], aggfunc={'D':np.mean,'E':[np.sum,max,min]}, values=['D','E'])
```
