```python
img_grey=img.sum(2) / (255*3)
```

```python
fig = plt.figure(figsize=(7,7))

img_grey = 255*3 - img_grey # here we reduced  color
fig.add_subplot(1,2,1)  # 1 is for first row and 2 is for 2 chart and 1 is combine chart (here we combined 2 chart in same row)
plt.imshow(img_grey)

img = 255- img
fig.add_subplot(1,2,2)
plt.imshow(img)
```
