```python
create database company;

use company;

## Create Employee Table
Create table Employee(
Emp_Id int not null primary key auto_increment,
First_Name varchar(25),
Last_Name varchar(25),
Salary float(15),
Joining_Date datetime,
Department_Name varchar(25)
);

## Check the schema
desc Employee;

## Create Bonus Table
Create table Bonus(
Emp_Ref_Id int,
Bonus_Amt float,
Bonus_Date datetime,
foreign key(Emp_Ref_Id) references Employee(Emp_Id)
on Delete Cascade
);

## Create Employee Designation Table
Create table Designation(
Emp_Ref_Id int,
Designation varchar(25),
Designation_Date datetime,
foreign key(Emp_Ref_Id) references Employee(Emp_Id)
on Delete Cascade
);

insert into Employee values
(0001, 'Krish', 'Naik','50000', '14-12-11 09.00.00', 'Development'),
(0002, 'Sudanshhu', 'Kumar', '60000', '14-12-11 09.00.00', 'Development'),
(0003, 'Sanketh', 'Kumar', '70000', '14-12-11 09.00.00', 'HR'),
(0004, 'Darius', 'Bengali', '30000', '15-12-11 09.00.00', 'Accountant'),
(0005, 'Shaktiman', 'Hero', '50000', '15-12-11 09.00.00', 'Accountant');

select * from Employee;

insert into Bonus values
(0001, 5000, '16-03-14'),
(0002, 5000, '16-03-13'),
(0003, 5000, '16-03-15'),
(0003, 5000, '16-03-17'),
(0004, 3500, '16-03-19'),
(0005, 700, '16-03-20'),
(0001, 8000, '16-03-21'),
(0002, 8500, '16-03-21');

select * from Bonus;
```
