#### Comparision Operator:  
                           --->   <, >, <=, >=, !=, ==

```python
num1=10
num2=5

print(num1 == num2)          # ==  opeator compare the two values.
print(num1 > num2)           # It will tell the True or False.
print(num1 < num2)
print(num1 != num2)
```

#### Logical Operators :   
                        ---> and , or, not

```python
print(10==11 and 20==20)   #  ---> and operator tell if any one of the operands is false then output will be false.
print(10==11 or 20==20)    #  ---> or operator tell if any both operands is false then output will be false.
print(not(10==11))         #  ---> not operator will tell if both operands are true or false then the output will be false.
```

#### Membership Operators:    
                            --->   in , not in operaotr

```python
l=[1,2,3,4,5,6,7]
print(4 in l)
print(8 in l)
print (9 not in l)

s="Python String"
print("python" in s)    # membership operators is a case sensitive 
print("Python" in s)      

```
